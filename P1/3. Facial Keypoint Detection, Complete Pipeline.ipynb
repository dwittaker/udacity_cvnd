## Face and Facial Keypoint detection

After you've trained a neural network to detect facial keypoints, you can then apply this network to *any* image that includes faces. The neural network expects a Tensor of a certain size as input and, so, to detect any face, you'll first have to do some pre-processing.

1. Detect all the faces in an image using a face detector (we'll be using a Haar Cascade detector in this notebook).
2. Pre-process those face images so that they are grayscale, and transformed to a Tensor of the input size that your net expects. This step will be similar to the `data_transform` you created and applied in Notebook 2, whose job was tp rescale, normalize, and turn any iimage into a Tensor to be accepted as input to your CNN.
3. Use your trained model to detect facial keypoints on the image.

Once you have an image to work with (and, again, you can select any image of faces in the `images/` directory), the next step is to pre-process that image and feed it into your CNN facial keypoint detector.

First, load your best model by its filename.

```python
import torch
from models import Net
import torchvision.transforms.functional as TF
import random

# def rotateit(image, key_pts):
#         angle = cv2.ROTATE_90_CLOCKWISE # random.choice(self.angles)
        
#         #image, key_pts = sample['image'], sample['keypoints']
        
#         image_copy = np.copy(image.detach())
#         key_pts_copy = np.copy(key_pts.detach())
#         #print(image_copy)
#         #print(key_pts_copy)
#         rotimg = cv2.rotate(image_copy, angle)
#         rotkpt = cv2.rotate(key_pts_copy, angle)
#         #print(rotimg)
#         #print(rotkpt)
        
#         return rotimg, rotkpt



def rescale(image, outsize):
    h, w = image.shape[:2]
    if isinstance(outsize, int):
        if h > w:
            new_h, new_w = outsize * h / w, outsize
        else:
            new_h, new_w = outsize, outsize * w / h
    else:
        new_h, new_w = outsize

    new_h, new_w = int(new_h), int(new_w)

    img = cv2.resize(image, (new_w, new_h))
    return img

def converttotensor(image):
    #image, key_pts = sample['image'], sample['keypoints']

    # if image has no grayscale color channel, add one
    if(len(image.shape) == 2):
        # add that third color dim
        image = image.reshape(image.shape[0], image.shape[1], 1)

    # swap color axis because
    # numpy image: H x W x C
    # torch image: C X H X W
    image = image.transpose((2, 0, 1))
    image = torch.from_numpy(image)
    image = image.type(torch.FloatTensor)
    
    return image

def load_ckp(checkpoint_fpath, model):
    checkpoint = torch.load(checkpoint_fpath , map_location=lambda storage, loc: storage)
    revised_state_dict = create_revised_state_dict(model.state_dict(), checkpoint['state_dict'])
    model.load_state_dict(revised_state_dict)
    return model, checkpoint['epoch'], checkpoint['batchloss']

#taken from https://discuss.pytorch.org/t/how-to-load-part-of-pre-trained-model/1113/2

def create_revised_state_dict(model_dict, state_dict):
 
    own_state = model_dict
    for name, param in state_dict.items():
        if name not in own_state:
             continue
        #if isinstance(param, Parameter):
            # backwards compatibility for serialized parameters
        param = param.data
        own_state[name].copy_(param)
    return own_state

def net_predict_output(testimage):

    global net
    
    # forward pass to get net output
    output_pts = net(testimage)

    # reshape to batch_size x 68 x 2 pts
    output_pts = output_pts.view(output_pts.size()[0], 68, -1)
    #print(output_pts)
    return testimage, output_pts

def show_all_keypoints(image, predicted_key_pts):
    """Show image with predicted keypoints"""
    # image is grayscale
    #print(image.shape)
    #print(predicted_key_pts.shape)
    
    plt.imshow(image, cmap='gray')
    plt.scatter(predicted_key_pts[:,:, 0], predicted_key_pts[:,:, 1], s=20, marker='.', c='m')
    # plot ground truth points as green pts
    #print(predicted_key_pts[:, 0], predicted_key_pts[:, 1])

def visualize_output(test_image, test_output):
    batch_size = 1
    for i in range(batch_size):
        plt.figure(figsize=(10,5))
        ax = plt.subplot(1, batch_size, i+1)

        # un-transform the image data
        image = test_image   # get the image from it's Variable wrapper
        if str(type(image)) == "<class 'torch.Tensor'>":
            image = image.detach() 
            image = image.cpu()
            image = image.numpy()   # convert to numpy array from a Tensor
            image = np.squeeze(image, axis=0)
            image = np.transpose(image, (1, 2, 0))   # transpose to go from torch to numpy image

        # un-transform the predicted key_pts data
        #print(type(test_output))
        predicted_key_pts = test_output
        if str(type(predicted_key_pts)) == "<class 'torch.Tensor'>":
            predicted_key_pts = predicted_key_pts.detach()
            predicted_key_pts = predicted_key_pts.cpu()
            predicted_key_pts = predicted_key_pts.numpy()
        # undo normalization of keypoints  
        predicted_key_pts = predicted_key_pts*50.0+100
        
        
        # call show_all_keypoints
        show_all_keypoints(np.squeeze(image), predicted_key_pts)
        #print(predicted_key_pts)  
        plt.axis('off')

    
    plt.show()




net = Net()

## TODO: load the best saved model parameters (by your path name)
## You'll need to un-comment the line below and add the correct name for *your* saved model
#net.load_state_dict(torch.load('saved_models/checkpoint.pt'))
#ckp_path = 'saved_models/checkpoint_naimish_ext_DW_0035.pt'
ckp_path = 'saved_models/checkpoint_0031_naimishbatch_DW.pt'


net,epoch,loss = load_ckp(ckp_path, net)
net = net.float()
## print out your net and prepare it for testing (uncomment the line below)
print(net)
print("Training Epochs: ", epoch, " Loss: ", loss)
net.eval()

```

### TODO: Transform each detected face into an input Tensor

You'll need to perform the following steps for each detected face:
1. Convert the face from RGB to grayscale
2. Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
3. Rescale the detected face to be the expected square size for your CNN (224x224, suggested)
4. Reshape the numpy image into a torch image.

**Hint**: The sizes of faces detected by a Haar detector and the faces your network has been trained on are of different sizes.
```python
image_copy = np.copy(image)

# loop over the detected faces from your haar cascade
for (x,y,w,h) in faces:
    
    # Select the region of interest that is the face in the image 
    roi = image_copy[y:y+h, x:x+w]
    #print(roi)
    
    ## TODO: Convert the face region from RGB to grayscale
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)  
    ## TODO: Normalize the grayscale image so that its color range falls in [0,1] instead of [0,255]
    gray =  gray/255.0
    
    ## TODO: Rescale the detected face to be the expected square size for your CNN (224x224, suggested)
    small = rescale(gray, 224)
    #print(small.shape)
    ## TODO: Reshape the numpy image shape (H x W x C) into a torch image shape (C x H x W)
    tensorimage = converttotensor(small)
    y = torch.unsqueeze(tensorimage, 0)
    #print(y.size())
    ## TODO: Make facial keypoint predictions using your loaded, trained network 
    test_images, test_outputs = net_predict_output(y)
    #print(test_images.data.size())
    #print(test_outputs.data.size())
    ## TODO: Display each detected face and the corresponding keypoints 
    #print(test_outputs[:,:,1])
    visualize_output(test_images, test_outputs)
```
